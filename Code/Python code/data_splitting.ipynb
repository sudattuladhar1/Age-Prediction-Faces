```python
import numpy as np
import pandas as pd
from PIL import Image
import shutil
```

```python
data = pd.read_csv('data_preprocessed.csv').iloc[:,2:]
data
```

```python
data = data.sample(frac=1)
data
```

```python
Nd = len(data)
Data_train = data.iloc[0:int(.6*Nd)]
Data_validation = data.iloc[int(.6*Nd):int(0.8*Nd)]
Data_test = data.iloc[int(0.8*Nd):Nd]

print(f"All Data included: {len(Data_train) + len(Data_validation) + len(Data_test) == len(data)}")
```

```python
def datacopy(df,dest_directory):
    for i in range(len(df)):
        jpgfile = f"data_preprocessed/{df.iloc[i]['filename']}"
        shutil.copy(jpgfile, dest_directory)
    
```

```python
N_Train = 10000
N_Valid = 1000
N_Test = 1000
datacopy(Data_train.iloc[:N_Train], 'Training_data');
datacopy(Data_validation.iloc[:N_Valid], 'Validation_data');
datacopy(Data_test.iloc[:N_Test], 'Testing_data');
```

```python
Data_train.iloc[:N_Train].to_csv('Data_train.csv')
Data_validation.iloc[:N_Valid].to_csv('Data_validation.csv')
Data_test.iloc[:N_Test].to_csv('Data_test.csv')
```

```python
import os.path
def TestFileExists(df, dest_dir):
    for i in range(len(df)):
        jpgfile = f"{dest_dir}/{df.iloc[i]['filename']}"
        if not os.path.exists(jpgfile):
            print(jpgfile)
```

```python
TestFileExists(Data_test, 'Testing_data')
```

```python
TestFileExists(Data_validation, 'Validation_data')
```

```python
TestFileExists(Data_train, 'Training_data')
```
