```python
import numpy as np
import pandas as pd
#import cv2
from PIL import Image
```

```python
data = pd.read_csv('data_filtersize.csv')
```

```python
data['filename'] = None
data
```

```python
for i in range(len(data)):
#for i in range(10):
    filename = data['full_path'].iloc[i]
    img = Image.open(filename).convert('L')
    img = img.resize((64,64))
    
    filename_dest = f"img_{i}.jpg"
    
    img.save(f"data_preprocessed/{filename_dest}") # File created at new location
    data['filename'].iloc[i] = filename_dest
        
    if i % 1000 == 0:
        print('file#', i)
```
